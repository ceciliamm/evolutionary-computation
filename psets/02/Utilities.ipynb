```python
# F1 Plot Generator
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('tableau-colorblind10')
print(plt.style.available)

def f1(x):
    return x**2 - (2 * math.e * x) + (math.e**2) - 2

def f2(x):
    """F2 implementation."""
    return ((np.e - x) ** 6) - 6


def plot_f1(X, Y):
    plt.figure()
    plt.plot(X, Y)
    plt.ylabel("f1(x)")
    plt.xlabel("x")
    plt.plot(2.718281828459045, f1(2.718281828459045),'ro')
    plt.show()

X = np.linspace(-10, 15, num=100)
Y = f1(X)
plot_f1(X, Y)

X2 = np.linspace(-.4, 5.8, num=100)
Y2 = f2(X2)
plot_f1(X2, Y2)
```
