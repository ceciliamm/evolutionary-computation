```python
# F1 Plot Generator
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
    return x**2 - (2 * math.e * x) + (math.e**2) - 2

def f2(x):
    """F2 implementation."""
    return ((np.e - x) ** 6) - 6


def plot_f(X, Y, opt):
    plt.figure()
    plt.plot(X, Y)
    plt.ylabel("f1(x)")
    plt.xlabel("x")
    plt.plot(opt, f1(opt),'ro')
    plt.show()

X = np.linspace(-10, 15, num=100)
Y = f1(X)
plot_f(X, Y, 2.718281828459045)

X2 = np.linspace(-.4, 5.8, num=100)
Y2 = f2(X2)
plot_f(X2, Y2, 2.60939515596596)
```
