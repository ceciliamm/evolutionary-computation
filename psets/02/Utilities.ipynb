```python
# F1 Plot Generator

import math
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('tableau-colorblind10')
print(plt.style.available)

def f1(x):
    return x**2 - (2 * math.e * x) + (math.e**2) - 2


def plot_f1(X, Y):
    plt.figure()
    plt.plot(X, Y)
    plt.ylabel("f1(x)")
    plt.xlabel("x")
    plt.savefig('/Users/pablotrinidad/Desktop/f1.png', dpi=150)

X = np.linspace(-10, 15, num=100)
Y = f1(X)
plot_f1(X, Y)
```
